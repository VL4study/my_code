```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score
```

```python
data = pd.read_csv('data-logistic.csv', header=None)
data.head()
```

```python
y = data.iloc[:, 0].to_numpy()
X = data.iloc[:, 1:].to_numpy()
```

```python
s = np.array([1, 3])
```

```python
s[0] = s[0] + 1
```

```python
s[0]
```

```python
def gradient_descent(X, y, alpha=0.1, Reg=None, C=10):
    l = X.shape[0]
    n = X.shape[1]
    w_prev = np.array([0.0] * n)
    w = np.array([0.0] * n)
    if Reg:
        if Reg == 'L2':
            _reg = lambda x: x
    else:
        _reg = lambda x: 0
    step = 0
    eps = 1e3
    while eps > 1e-5 and n != 10000:
        for j in range(n):
            w_prev[j] = w[j]
            w[j] = w[j] + alpha * (1 / l) * sum([y[i] * X[i, j] * (1 - 1 / (1 + np.exp(-y[i] * (sum([w[k] * X[i, k] for k in range(n)]))))) for i in range(l)]) - alpha * C * _reg(w[j]) 
        eps = (sum([(w[k] - w_prev[k])**2 for k in range(n)]))**(1/2)
        step += 1
    return w    
```

```python
w = gradient_descent(X, y)
```

```python
y_pred = [sum([w[j] * X[i, j] for j in range(X.shape[1])]) for i in range(X.shape[0])]
```

```python
sigma = lambda x: 1 / (1 + np.exp(-x))
y_pred_prob = np.array([sigma(y) for y in y_pred])
y_pred_prob
```

```python
score1 = round(roc_auc_score(y, y_pred_prob), 3)
```

```python
w = gradient_descent(X, y, Reg='L2')
```

```python
y_pred = [sum([w[j] * X[i, j] for j in range(X.shape[1])]) for i in range(X.shape[0])]
y_pred_prob = np.array([sigma(y) for y in y_pred])
y_pred_prob
```

```python
score2 = round(roc_auc_score(y, y_pred_prob), 3)
```
